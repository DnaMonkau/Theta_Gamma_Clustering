```python
def get_data(region,mc):
    '''
    Process the REM sleep data for each trial and admit them into a dataframe
    input: region and control or mutation
    output: dataframe containing rat number, condition, mutation/control, 
    epoch number, region, and the rem sleep data s an array
     
    '''
    datan = np.array([])
    dataframe = df(columns = ['Rat', 'Condition', 'MC','Epoch','Region', 'REM Data'])
    dataframe['REM Data'] = dataframe['REM Data'].astype(object)
    if region == 'PFC':
        if mc=='Control':
            states = {1:PFC_states_, 2:PFC_states_2, 6:PFC_states_6, 9:PFC_states_9}
            data_ = [PFC_data, PFC_data2, PFC_data6, PFC_data9]
        if mc=='Mutation':
            states = {3:PFC_states_3,4: PFC_states_4, 7:PFC_states_7, 8:PFC_states_8}
            data_ = [PFC_data3, PFC_data4, PFC_data7, PFC_data8]
    if region == 'HPC':
        if mc=='Control':
            states = {1:states_, 2:states_2, 6:states_6, 9:states_9}
            data_ = [datas, datas2, datas6, datas9]
        if mc=='Mutation':
            states = {3:states_3,4: states_4, 7:states_7, 8:states_8}
            data_ = [datas3, datas4, datas7, datas8]
    j=0
    
    for s, d in zip(states.keys(), range(len(data_))):
        for state, data in zip(states[s], data_[d]):
            plots=[]
            epoch=[]
            j+=1
            x= np.where(state==5, True, False)[0]
            start = 0
            y = 1000
            for i in range(len(x)-1):
                if x[i]:
                    datan = np.append(datan, data[start:y]*x[i])
                    if not x[i+1]:
                        epoch.append(datan)
                        datan = np.array([])
                start = y
                y = start+1000
            e=1
            for dt in epoch:
                x =  dt
#                 plot(x)
#                 xlim([2000,4000])
#                 show()
                dataframe  = dataframe.append({'Rat': s, 'Condition': 'HomeCage', 'MC':mc,'Epoch': e,'Region':region,\
                                                           'REM Data':x.tolist()}, ignore_index=True)
                e+=1
    return dataframe
```
